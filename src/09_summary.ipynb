# Сводка результатов курсовой работы

Ноутбук собирает таблицы результатов из папки `results` и показывает лучшие модели по каждой задаче.


```python
from pathlib import Path
import pandas as pd

RESULTS_DIR = Path('../results')
files = sorted(RESULTS_DIR.glob('*_metrics.csv'))
rows = []
for path in files:
    df = pd.read_csv(path)
    if 'RMSE' in df.columns:
        best = df.sort_values('RMSE').iloc[0].to_dict()
        metric = 'RMSE'
        score = best['RMSE']
    else:
        best = df.sort_values('F1', ascending=False).iloc[0].to_dict()
        metric = 'F1'
        score = best['F1']
    rows.append({
        'file': path.name,
        'best_model': best['model'],
        'main_metric': metric,
        'score': score,
    })

summary = pd.DataFrame(rows)
display(summary)
summary.to_csv(RESULTS_DIR / 'summary_from_notebooks.csv', index=False)
```

## Общий вывод

В работе были решены 7 задач: три задачи регрессии и четыре задачи классификации. Для каждой задачи сравнивались несколько классических моделей машинного обучения и выполнялся подбор гиперпараметров для GradientBoosting. В качестве защиты от утечки данных из признаков удалялись `IC50, mM`, `CC50, mM` и `SI`.
